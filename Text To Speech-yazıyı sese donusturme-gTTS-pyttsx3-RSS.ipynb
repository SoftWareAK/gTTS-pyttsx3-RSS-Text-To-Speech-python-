#### Voice Speaking Rate
Pyttsx3 is a computer-generated human voice that sometimes can be robotic because humans not always speak slowly. We can increase the speed of speaking by setting the rate.

```python
import pyttsx3
engine = pyttsx3.init() # object creation

""" RATE setting"""
rate = engine.getProperty('rate')  
print(rate)                        
engine.setProperty('rate', 150)
```

#### Increase Volume

```python
import pyttsx3
engine = pyttsx3.init()

"""Setting Volumn"""
#get current volumn level
volume = engine.getProperty('volume')
print (volume)

#setting volumn
engine.setProperty('volume',1.0)  
engine.say("I will speak this text")
engine.runAndWait()
```

The volume values are between 0 and 1 which means you can’t go above the “1” instead you can set 0.5, 0.4 or etc. So the minimum volume is 0 and maximum volume is 1.

# Automatically converting the news we scraped from the internet with RSS and gTTs into text and mp3 files

 RSS genellikle haber siteleri tarafından kullanılan, yeni eklenen içeriği kolaylıkla takip etmeye yarayan bir web formatıdır. Genellikle .xml veya .rss uzantılı dosyalardan oluşmaktadır. Birçok haber sitesinde bu özellikle karşılaşabilirsiniz. 

```python
import requests
import xml.etree.ElementTree as elementTree
from gtts import gTTS


def load_rss(url):
    try:
        response = requests.get(url)
        file = open('news_rss.xml', 'wb')
        file.write(response.content)
    except:
        print('An error occurred')


def parse_xml(xmlFile):
    xml = elementTree.parse(xmlFile)
    root = xml.getroot()

    news_items = []

    for item in root.findall('./channel/item'):
        news = {}

        for child in item:
            if child.tag == 'http://search.yahoo.com/mrss/}content':
                news['media'] = child.attrib['url']
            else:
                if child.text is not None:
                    news[child.tag] = child.text.encode('utf-8')
        news_items.append(news)

    return news_items


def create_news_file(news_items):
    for item in news_items:
        description = item.get('description').decode('utf-8')
        index = str(description).find('</a>')
        file_name = item.get('title').decode('utf-8')[0:10].replace('"', '')
        file = open(file_name + '.txt', 'w+')
        file.write(description[index + 4:] + '\n\n')
        file.close()
        file = open(file_name + '.txt', 'r')
        text_to_read = file.read()
        tts = gTTS(text_to_read, lang='tr')
        tts.save(file_name + '.mp3')


load_rss('http://www.haberturk.com/rss')
news_items = parse_xml('news_rss.xml')
create_news_file(news_items)
print('Created files')
```
